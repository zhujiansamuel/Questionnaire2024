```python
import time
time_s = time.time()
from datetime import date
surveys = Survey.objects.filter(
            is_published=True, expire_date__gte=date.today(), publish_date__lte=date.today()
        ).prefetch_related("responses")
for survey in surveys:
    responses = Response.objects.filter(survey=survey).exists()
    print(responses)
time_e = time.time()
print((time_e-time_s)*1000)
```

```python
import time
time_avg = []
for i in range(60):
    time_s = time.time()
    from datetime import date
    surveys = Survey.objects.filter(
                is_published=True, expire_date__gte=date.today(), publish_date__lte=date.today()
            ).prefetch_related("responses")
    for survey in surveys:
    #     responses = Response.objects.filter(survey=survey).exists()
        print(survey.responses.exists())
    time_e = time.time()
    time_avg.append(time_e - time_s)
print(sum(time_avg) / len(time_avg)*1000 )
```
