```python
import csv
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
filepath = '学生データベース.csv'
filepath_1 = '第１回_イントロダクション～人工知能と認知科学フィードバック.csv'
filepath_2 = '第2回_認知科学の考え方～情報処理の三水準.csv'
filepath_3 = '第３回_認知科学の方法.csv'
filepath_4 = '第４回_学術論文の構成 と 議論の進め方.csv'
filepath_5 = '第５回_学術論文を読む.csv'
```

```python
df = pd.read_csv(filepath)
df = df.drop(['Unnamed: 0', 'Unnamed: 2'], axis=1, errors='ignore')
df_1 = pd.read_csv(filepath_1)
df_2 = pd.read_csv(filepath_2)
df_3 = pd.read_csv(filepath_3)
df_4 = pd.read_csv(filepath_4)
df_5 = pd.read_csv(filepath_5)
```

```python
df
```

```python
columns_name = ["第１回","第2回","第３回","第４回","第５回"]
df_list = [df_1,df_2,df_3,df_4,df_5]
for index, df_s in enumerate(df_list):
  for row_1 in df_s.itertuples():
    insert = True
    for row in df.itertuples():
      # print(getattr(row_1,"学生番号"),"------",getattr(row, "学籍番号"))
      # print(type(getattr(row_1,"学生番号")))
      # print(type(getattr(row, "学籍番号")))
      if not pd.isna(getattr(row_1,"学生番号")):
        if getattr(row_1,"学生番号") == str(getattr(row, "学籍番号")):
          print(getattr(row_1,"学生番号"),"------",getattr(row, "学籍番号"))
          df.at[row.Index,columns_name[index]] = 1
          insert = False
    if insert:
      print(getattr(row_1,"学生番号"))
      df = pd.concat([df, pd.DataFrame({'学籍番号':[getattr(row_1,"学生番号")],columns_name[index]:1})], ignore_index=True)
```

```python
present_list = []
for row in df.itertuples():
  print(row)
  if not pd.isna(getattr(row, "_5")):
    present_list.append(str(getattr(row, "学籍番号")))
present_list
```

```python
from pptx import Presentation
from pptx.util import Inches, Pt

prs = Presentation()

title_slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]

title.text = "認知科学概論 \n第６回 学術論文の要約報告会"
subtitle.text = "発表会"

slide.shapes.title.text_frame.paragraphs[0].font.size = Pt(55)

for index,id in enumerate(present_list):
  
  title_slide_layout = prs.slide_layouts[0]
  slide = prs.slides.add_slide(title_slide_layout)
  title = slide.shapes.title
  subtitle = slide.placeholders[1]
  
  
  if index+1 != len(present_list):
    title.text = "発表者\n\n" + id + "     \n\n ( "+ str(index+1) + "/" + str(len(present_list)) +" )"
    subtitle.text = " \n \n \n次の発表者：" + present_list[index+1]
    slide.shapes.title.text_frame.paragraphs[0].font.size = Pt(65)
  else:
    title.text = "発表者：" + id + "    \n ( "+ str(index+1) + "/" + str(len(present_list)) +" )"
    subtitle.text = " \n \n \n(最後の発表者)"
    slide.shapes.title.text_frame.paragraphs[0].font.size = Pt(65)

title_slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]

slide.shapes.title.text_frame.paragraphs[0].font.size = Pt(55)

title.text = "本日の発表は完了しました。"
subtitle.text = "ありがとうございます。!"



prs.save('test.pptx')
```
